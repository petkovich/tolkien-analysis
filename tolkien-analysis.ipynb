```python
import pandas as pd
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
dosave = False
```

```python
df = pd.read_csv("data.csv");
```

```python
ax = df["Godina studija"].plot.hist(rwidth=.8, bins=np.arange(min(df["Godina studija"]), max(df["Godina studija"])+2) - 0.5)
ax.set_xlim((.5,6.5))

ax.set_title("Number of students per year of study")
ax.set_xlabel("Year of study")

ax.set_ylabel("Number of students");
if dosave:
    plt.savefig('figs/eps/students.eps', format='eps')
    plt.savefig('figs/png/students.png', format='png', dpi=1200)

print( df["Godina studija"].value_counts());
```

```python
answers_id = df.columns[7:-1]
mydata = df[answers_id]
col = df.loc[:,df.columns[7]:df.columns[-1]]
df['avg_grade'] = col.mean(axis=1)
all_grades = []


myselection = df
answers_id = myselection.columns[7:-2]
mydata = myselection[answers_id]
mydata=mydata.sort_index(axis=1)
#print(mydata)
for item in answers_id:
    #print(item)
    #print(item, mydata[item].mean())
    #print( mydata[item][:-2])
    #mydata[item].hist()
    all_grades+= list(mydata[item])

[print("Number of " + str(i) +": " + str(all_grades.count(i))) for i in range(1,6)]
    
plt.axvline(df["avg_grade"].mean(), color='k', linestyle='dashed', linewidth=1)
plt.hist(all_grades, rwidth=.8, bins=np.arange(min(df['avg_grade']), max(df['avg_grade'])+2) - 0.5)
plt.xlim((.5,5.5))
plt.text(df['avg_grade'].mean()*0.5,max(plt.ylim())*0.9, 'Average grade: {:.2f}'.format(df['avg_grade'].mean()))

plt.title("Grade distribution")
plt.xlabel("Grade")
plt.ylabel("Number of grades");
#plt.savefig('figs/eps/grades.eps', format='eps')
if dosave:
    plt.savefig('figs/eps/grades.eps', format='eps')
    plt.savefig('figs/png/grades.png', format='png', dpi=1200)
```

```python
lc =len(prevoditelji_chad[0])
lp =len(prevoditelji_pleb[0])
myweights = [[1000/lc]*lc, [1000/lp]*lp]
```

```python
prvi = [prevoditelji_chad[0], prevoditelji_pleb[0]]
ax = plt.hist(prvi, rwidth=.8, bins=np.arange(1, 5+2) - 0.5, label = ("Education ", "No Education "),  weights = (myweights))
plt.legend()

ed_av = sum(prvi[0])/len(prvi[0])
uned_av = sum(prvi[1])/len(prvi[1])
plt.axvline(ed_av, color='b', linestyle='dashed', linewidth=1)
plt.axvline(uned_av, color='r', linestyle='dashed', linewidth=1)
plt.ylim(0,300)
plt.text(ed_av*1.02,max(plt.ylim())*0.9, '{:.2f}'.format(ed_av), color='b')
plt.text(uned_av*0.85,max(plt.ylim())*0.9, '{:.2f}'.format(uned_av), color='r')


plt.title('Maras\' translation grade distribution')
plt.ylabel("Number of grades")
plt.xlabel("Grade");
#plt.savefig('figs/maras.eps', format='eps')
if dosave:
    plt.savefig('figs/eps/maras.eps', format='eps')
    plt.savefig('figs/png/maras.png', format='png', dpi=1200)
[print("Total Number of Education " + str(i) +": " + str(prvi[0].count(i))) for i in range(1,6)]
[print("Total Number of No Education " + str(i) +": " + str(prvi[1].count(i))) for i in range(1,6)]
[print("Normalized Number of Education " + str(i) +": " + str(prvi[0].count(i)*myweights[0][0])) for i in range(1,6)]
[print("Normalized Number of No Education " + str(i) +": " + str(prvi[1].count(i)*myweights[1][0])) for i in range(1,6)];
```

```python
drugi = [prevoditelji_chad[1], prevoditelji_pleb[1]]
ax = plt.hist(drugi, rwidth=.8, bins=np.arange(1, 5+2) - 0.5, label = ("Education ", "No Education "), weights = (myweights))
plt.legend()
ed_av = sum(drugi[0])/len(drugi[0])
uned_av = sum(drugi[1])/len(drugi[1])
plt.axvline(ed_av, color='b', linestyle='dashed', linewidth=1)
plt.axvline(uned_av, color='r', linestyle='dashed', linewidth=1)
plt.ylim(0,350)
plt.text(ed_av*1.02,max(plt.ylim())*0.9, '{:.2f}'.format(ed_av), color='b')
plt.text(uned_av*0.85,max(plt.ylim())*0.9, '{:.2f}'.format(uned_av), color='r')
plt.title('Crnković\'s translation grade distribution')
plt.ylabel("Number of grades")
plt.xlabel("Grade");
#plt.savefig('figs/crnkovic.eps', format='eps')
if dosave:
    plt.savefig('figs/eps/crnkovic.eps', format='eps')
    plt.savefig('figs/png/crnkovic.png', format='png', dpi=1200)
[print("Total Number of Education " + str(i) +": " + str(drugi[0].count(i))) for i in range(1,6)]
[print("Total Number of No Education " + str(i) +": " + str(drugi[1].count(i))) for i in range(1,6)]
[print("Normalized Number of Education " + str(i) +": " + str(drugi[0].count(i)*myweights[0][0])) for i in range(1,6)]
[print("Normalized Number of No Education " + str(i) +": " + str(drugi[1].count(i)*myweights[1][0])) for i in range(1,6)];
```

```python
from scipy.stats import ttest_ind, ttest_rel
from numpy import mean, var

print("Maras: " + str(ttest_ind(prvi[0], prvi[1])))
print(mean(prvi[0]), var(prvi[0]))
print(mean(prvi[1]),  var(prvi[1]))

print("Crnković: " + str(ttest_ind(drugi[0], drugi[1])))
print(mean(drugi[0]), var(drugi[0]))
print(mean(drugi[1]),  var(drugi[1]))
educirani = []
needucirani = []
for i in range(len(prvi[0])):
    educirani.append(prvi[0][i]/drugi[0][i])
    needucirani.append(prvi[1][i]/drugi[1][i])
    
print("Podijeljeno: " + str(ttest_rel(educirani, needucirani))) #rel/ind, dijeljenje
print(mean(educirani), var(educirani))
print(mean(needucirani), var(needucirani))
    
    
```
